```python
# list of initial options
options = {
    1: "Spikes",
    2: "Seizures",
    3: "SWDs",
    4: "Spindles"
}

# Set default as false
spikes = 'FALSE'
seizures = 'FALSE'
swds = 'FALSE'
spindles = 'FALSE'

# User input to select options
print("Select your options:")
for key, value in options.items():
    print(f"{key}. {value}")

class SelectedOptions:
    pass

selected_options = SelectedOptions()

# Initialize variables for Threshold and Rolling selection
threshold_selected = False
rolling_selected = False

while True:
    choice = input("Enter the option number (0 to finish): ")

    if choice == "0":
        break

    choice = int(choice)
    if choice == 1:  # For "Spikes" option
        spikes_selected = False

        print("Select additional options for Spikes:")
        while True:
            print("Enter both a and b to use both methods")
            print("a. Threshold")
            print("b. Rolling")
            spikes_choice = input("Enter the option number (0 to finish Spikes selection): ")
            spikes_choice = spikes_choice.strip().lower()

            if spikes_choice == "0":
                break
            elif spikes_choice == "a":
                threshold_selected = True
                spikes_selected = True
            elif spikes_choice == "b":
                rolling_selected = True
                spikes_selected = True

        if spikes_selected:
            setattr(selected_options, "spikes", True)

    elif choice == 4:  # For "Spindles" option
        setattr(selected_options, "spindles", True)

    elif choice in options:
        option_name = options[choice]
        setattr(selected_options, option_name.replace(" ", "_").lower(), True)

# Tell the user what they picked
print("You have chosen to screen for:")
for option in options.values():
    if hasattr(selected_options, option.replace(" ", "_").lower()):
        print(option)

# Check if Threshold or Rolling were selected and print the appropriate message
if threshold_selected:
    print("Threshold method of spike counting was selected.")
if rolling_selected:
    print("Rolling method of spike counting was selected.")

# Set new variables (spikes, seizures, swds, spindles, spikes_threshold, spikes_rolling)
# as True or False depending on user input
for option in options.values():
    variable_name = option.replace(" ", "_").lower()
    if hasattr(selected_options, variable_name):
        value = getattr(selected_options, variable_name)
        locals()[variable_name] = value
```
